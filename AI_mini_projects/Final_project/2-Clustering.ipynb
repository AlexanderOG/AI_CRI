```python
def cleanList(x):
  aux = x.split()
  aux[0] = aux[0][1:]
  aux[-1] = aux[-1][0:-1]
  aux = [float(i) for i in aux if i != '']
  return aux
```

```python
data["Encoded"] = data["Encoded"].apply(cleanList)
```

```python
split_df = pd.DataFrame(data['Encoded'].tolist())
split_df
```

```python
inertias = {}
K = [300, 400, 500, 600]
for k in K:
  kmeanModel = KMeans(n_clusters=k).fit(split_df)
  inertias[k] = kmeanModel.inertia_
```

```python
best_k = 300
kmeanModel = KMeans(n_clusters=best_k).fit(split_df)
```

```python
data["clusters"] = kmeanModel.labels_
```

```python
pd.set_option("display.max_colwidth", 140) #Important to show complete text of tweets
data.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
# put all sentence embeddings in a matrix
embed_mat = np.array([x for x in data["Encoded"]])
# calculate distance between every embedding pair
sim_mat = cosine_similarity(embed_mat,embed_mat)
```

```python
#get sim score for a given sentence at position df.iloc[sentence_id]
sentence_id = 300
print("Similarities for Sentence : " + data.iloc[sentence_id].Text)
# write sim scores to df
data['sim_score'] = sim_mat[sentence_id]
```

```python
#Bad solution
aux_reco = data.sort_values('sim_score')
aux_reco.head()
```

```python
#Good solution only shows values in the same cluster as the sentence
cluster = data.iloc[sentence_id].clusters
aux_reco = aux_reco[aux_reco["clusters"] == cluster]
aux_reco = aux_reco.sort_values('sim_score')
aux_reco.head()
```

```python
aux_reco = data.sort_values('sim_score', ascending=False)
aux_reco = aux_reco[aux_reco["sim_score"] < .99 ]
aux_reco.head(10)
```
